### Влияние  оценок пользователей и критиков на продажи внутри одной популярной платформы

```python
def corr_annotation(corr):
    if abs(corr) <= 0.3:
        return 'Очень слабая'
    elif abs(corr) <= 0.5:
        return 'Cлабая'
    elif abs(corr) <= 0.7:
        return 'Средняя'
    elif abs(corr) <= 0.9:
        return 'Сильная'
    elif abs(corr) <= 1:
        return 'Очень сильная'
    
print(corr_annotation(0.6))
```

```python
for i in df_new['platform'].unique():
    df_i = df_new.query('platform == @i')
    correlation_user = df_i['user_score'].corr(df_i['all_sales'])
    correlation_critic = df_i['critic_score'].corr(df_i['all_sales'])
    fig, ax = plt.subplots(1 ,2, figsize=(15,3))
    sns.regplot(x=df_i['user_score'], y=df_i['all_sales'], ax=ax[0],fit_reg=False)
    plt.suptitle(f'Зависимость оценок на продажи {i}')
    ax[0].set(ylabel='Продажи')
    ax[0].set(xlabel='Оценка пользователей')   
    sns.regplot(x=df_i['critic_score'], y=df_i['all_sales'], ax=ax[1],fit_reg=False)
    ax[1].set(xlabel='Оценка критиков')
    ax[1].set(ylabel='Продажи')
    plt.show()
    print(f' Корреляция пользовательских оценок и продаж равна: {correlation_user} : {corr_annotation(correlation_user)}')
    print(f' Корреляция оценок критиков и продаж равна: {correlation_critic} : {corr_annotation(correlation_critic)}')
```

### Влияние оценок на продажи игр вцелом

```python
corr_cr = df_new['critic_score'].corr(df_new['all_sales'])
sns.regplot(x=df_new['critic_score'],y=df_new['all_sales']);
print(f"Koppеляция равна {corr_cr} : {corr_annotation(corr_cr)}")
plt.title('Влияние оценок критиков на продажи')
plt.xlabel('Оценка критиков')
plt.ylabel('Продажи');
```

```python
corr_us = df_new['user_score'].corr(df_new['all_sales'])
sns.regplot(x=df_new['user_score'],y=df_new['all_sales']);
print(f"Koppеляция равна {corr_us} : {corr_annotation(corr_us)}")
plt.title('Влияние оценок пользователей на продажи')
plt.xlabel('Оценка критиков')
plt.ylabel('Продажи');
```

## Проверка гипотез  

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

```python
xone__user_score = df_new.query('platform == "XOne"').copy()
pc__user_score = df_new.query('platform == "PC"').copy()
```

Посмотрим статистические данные по Xbox и PC

```python
print(pc__user_score['user_score'].describe())
xone__user_score['user_score'].describe()
```

Вцелом данные очень похожи, так что установим не строгий порог для подтверждения теории.

Но вначале очистим данные от выбросов.

```python
def remove_outliers(data, columns):
    for column in columns:
        q_1,q_3 = data[column].quantile([0.25,0.75])
        iqr = q_3 - q_1
        low = q_1 - (1.5 * iqr)   
        up = q_3 + (1.5 * iqr)
        data = data[(data[column] <= up) & (data[column] >= low)]
    return data

#xone__user_score = remove_outliers(xone__user_score,['user_score'])
#pc__user_score = remove_outliers(pc__user_score, ['user_score'])
```

```python
# Комментарий ревьюера
print(xone__user_score.shape, df_new.query('platform == "XOne"').shape)
print(xone__user_score.user_score.mean(), df_new.query('platform == "XOne"').user_score.mean())
print(pc__user_score.shape, df_new.query('platform == "PC"').shape)
print(pc__user_score.user_score.mean(), df_new.query('platform == "PC"').user_score.mean())
```

H0- средние пользовательские оценки на PC и XOne одинаковые.  
H1-средние пользовательские оценки на PC и XOne разные.

```python
#В зависимость от желаемого результата выбираем alpha?
alpha= 0.01

result = st.ttest_ind(xone__user_score['user_score'].dropna(), pc__user_score['user_score'].dropna())

print(f'p-value = {result.pvalue}')

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу')
```

#### Вывод

Значение p-value слишком высоко, чтобы отвергнуть нулевую теорию. Так шанс случайно получить выборку с одинаковыми средними равен 14%, при установленном пороге в 1%.

### Средние пользовательские рейтинги жанров Action  и Sports разные.  

```python
action__user_score = df_new.query('genre == "Action"').copy()
sports__user_score = df_new.query('genre == "Sports"').copy()
```

Посмотрим статистические данные по Action и Sports

```python
print(action__user_score['user_score'].describe())
sports__user_score['user_score'].describe()
```

Очистим данные от редких и выбивающихся значений

```python
#action__user_score = remove_outliers(action__user_score,['user_score'])
#sports__user_score = remove_outliers(sports__user_score, ['user_score'])
```

H0- средние пользовательские оценки жанров Action и Sports одинаковые.  
H1-средние пользовательские оценки жанров Action и Sports разные.

```python
alpha = 0.01

result = st.ttest_ind(action__user_score['user_score'].dropna(), sports__user_score['user_score'].dropna())

print(f'p-value = {result.pvalue}')

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу')
```

#### Вывод

Значение p слишком мало, чтобы исключить случайность, так что отвергаем нулевую теорию о равенстве оценкок.
